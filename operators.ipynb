```python
#Assignment Operators
#1.Input a number and add 10 to it using +=.
a=int(input("enter first number"))
a+=10
print(a)
```

```python
#2.Input a number and subtract 5 from it using -=.
a=int(input("enter second number"))
a-=5
print(a)
```

```python
#3.Input a number and multiply it by 3 using *=.
a=int(input("enter first number"))
a*=3
print(a)
```

```python
#4 Input a number and divide it by 2 using /=.
a=int(input("enter first number"))
a/=2
print(a)
```

```python
#Comparison Operators
# 5.Input two numbers and check whether they are equal (==).
a=int(input("enter first number"))
b=int(input("enter second number"))
print(a==b)
```

```python
#6.Input two numbers and check whether the first number is greater than the second (>).
a=int(input("enter first number"))
b=int(input("enter second number"))
print(a>b)
```

```python
#7.Input two numbers and check whether the first number is less than or equal to the second (<=).
a=int(input("enter first number"))
b=int(input("enter second number"))
print(a<=b)
```

```python
#8.Input two numbers and check whether they are not equal (!=).
a=int(input("enter first number"))
b=int(input("enter second number"))
print(a!=b)
```

```python
#Logical Operators
# 9.Input your age and check if you are between 18 and 60 using and.
age=int(input("enter age"))
print=(age>=18 and age <=60)
```

```python
#10.Input a number and check if it is less than 0 or greater than 100 using or.
a=int(input("enter first number"))
print=(a<0 or a>100)
```

```python
#11.Input a boolean value (True or False) and print its opposite using not.
a=bool(input("enter true or false"))
print=not(a)
```

```python
#12.Input marks and attendance percentage. Check if marks are above 40 and attendance is above 75%.
marks=int(input("enter marks"))
attendance=int(input("enter attendance"))
print=(marks>40 and attendance>75)
```

```python
#Membership Operators
#13.Input a string and a character. Check whether the character is present in the string using in.
text=input("enter a string:")
char=input("enter a character:")
print(char in text)
```

```python
#14.Input a list of fruits and check whether "apple" is present in the list.
fruits=["apple","banana","orange","mango"]
print("apple" in fruits)
```
